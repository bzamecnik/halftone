# Examples of halftoning

```python
import halftone as ht
import PIL.Image
import numpy as np
```

```python
img = PIL.Image.open('../examples/lena.png')
```

```python
img
```

```python
ht.halftone(img, ht.euclid_dot(spacing=8, angle=30))
```

```python
gradient = ht.pil_from_array(np.tile(np.linspace(0, 1, 600).reshape(-1, 1), 300).T)
```

```python
gradient
```

```python
ht.halftone(gradient, ht.euclid_dot(spacing=30, angle=30))
```

```python
halftoned = ht.halftone(gradient, ht.euclid_dot(spacing=30, angle=30))
halftoned.save('gradient_halftone.png')
```

# Spot functions

- euclid_dot(angle, distance)
- square_dot(angle, distance)
- circle_dot(angle, distance)
- triangle_dot(angle, distance)
- line(angle, distance)

Angle is is degrees (CCW). Distance is the spacing of the grid.
